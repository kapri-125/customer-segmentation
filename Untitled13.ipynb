```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score, davies_bouldin_score
from scipy.cluster.hierarchy import dendrogram, linkage
df = pd.read_csv('Mall_Customers[1].csv')
df.head()
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]
inertia = []
k_range = range(1, 11)

for k in k_range:
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(X)
    inertia.append(model.inertia_)

plt.plot(k_range, inertia, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
k_opt = 5  # Example
kmeans = KMeans(n_clusters=k_opt, random_state=42)
k_labels = kmeans.fit_predict(X)
df['KMeans_Cluster'] = k_labels
linked = linkage(X, method='ward')

plt.figure(figsize=(10, 5))
dendrogram(linked)
plt.title('Dendrogram')
plhc = AgglomerativeClustering(n_clusters=5)
print(X.isnull().sum())
print("Silhouette Score (K-Means):", silhouette_score(X, k_labels))
print("Daviesâ€“Bouldin Index (K-Means):", davies_bouldin_score(X, k_labels))
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', hue='KMeans_Cluster', data=df, palette='Set2')
plt.title('Customer Segments - KMeans')
plt.show()
```
